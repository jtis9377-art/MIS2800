```python
%%sql
CREATE TABLE products (
    product_id INTEGER PRIMARY KEY,
    name TEXT,
    price REAL
) STRICT;
```

```python
%%sql
INSERT INTO products (name, price) VALUES
    ('Superbowl Jersey', 120.00),
    ('Team Hat', 35.00),
    ('Football', 60.00);
```

```python
%%sql
SELECT *
FROM products
WHERE price > 50;
```

```python
%%sql
UPDATE products
SET price = 99.99
WHERE name = 'Superbowl Jersey';
```

```python
%%sql
DELETE FROM products
WHERE product_id = 1;
```
