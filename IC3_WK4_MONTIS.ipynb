```python
%%sql
CREATE TABLE IF NOT EXISTS stock_performance(
    ticker TEXT,
    price_before_game REAL,
    price_week_after_game REAL,
    trading_volume_millions INTEGER
) STRICT;
```

```python
%%sql
-- create another table named ad_buys
CREATE TABLE IF NOT EXISTS ad_buys (
    ticker TEXT PRIMARY KEY,
    company_name TEXT,
    ad_cost_millions REAL,
    airtime_seconds INTEGER
) STRICT;
```

```python
%%sql
SELECT * FROM stock_performance;
```

```python
%%sql
SELECT ticker, price_before_game FROM stock_performance ORDER BY ticker ASC
```

```python
%%sql
SELECT * FROM stock_performance WHERE ticker LIKE 'T%'
```

```python
%%sql
SELECT ticker, price_week_after_game FROM stock_performance WHERE price_week_after_game > 150
```

```python
%%sql
SELECT ticker, company_name FROM ad_buys ORDER BY company_name DESC;
```
